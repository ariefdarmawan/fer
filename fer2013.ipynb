```python
from PIL import Image
import face_recognition as fr
import numpy as np
```

```python
imageid = 0
def image2Encoding(pixels, mode="RGB"):
    global imageid
    print("encoding data from image pixels. {}".format(imageid))
    imageid += 1
    imgdata = np.zeros((48,48,3), dtype=np.uint8)
    
    for index, pixel in enumerate(pixels):
        y = int(index % 48)
        x = int((index - y) / 48)
        #print("set data on x: {} y: {} v: {}".format(x,y,pixel))
        imgdata[x, y] = [pixel, pixel, pixel]
        
    #print(imgdata)
    img = Image.fromarray(imgdata, mode=mode)
    encodings = fr.face_encodings(imgdata)
    if len(encodings) > 0:
        return encodings[0]
    else:
        return []
```

```python
fer_training["encoding"] = fer_training["pixels_int"].apply(lambda x: image2Encoding(x))
fer_training.info()
```
